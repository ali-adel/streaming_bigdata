```python

daily_data = df.groupby('Formatted Date').agg({
    'Temperature (C)': 'mean',  
    'Apparent Temperature (C)': 'mean',
    'Humidity': 'mean',  
    'Wind Speed (km/h)': 'mean',  
    'Wind Bearing (degrees)': 'mean',
    'Visibility (km)': 'mean',
    'Loud Cover': 'mean',
    'Pressure (millibars)': 'mean',
    'Precip Type': 'max',
    
}).reset_index()
```

```python
daily_data.shape
```

```python
daily_data
```

```python
daily_data['Precip Type']=["snow" if code == 1 else "rain" for code in daily_data['Precip Type']]
```

```python
daily_data
```

```python
df['Loud Cover'].unique()
```

```python
df=daily_data
```

```python
df.drop(['Loud Cover'],inplace=True,axis=1)
```

```python
df.columns.to_list()
```

```python
df.rename(columns={'Formatted Date': 'Formatted_Date'}, inplace=True)
df.rename(columns={'Temperature (C)': 'Temperature'}, inplace=True)
df.rename(columns={'Wind Speed (km/h)': 'Wind_Speed'}, inplace=True)
df.rename(columns={'Visibility (km)': 'Visibility'}, inplace=True)
df.rename(columns={'Pressure (millibars)': 'Pressure'}, inplace=True)
df.rename(columns={'Precip Type': 'Precip_Type'}, inplace=True)
df.rename(columns={'Apparent Temperature (C)': 'Apparent_Temperature'}, inplace=True)
df.rename(columns={'Wind Bearing (degrees)': 'Wind_Bearing'}, inplace=True)
```

```python
df['Formatted_Date'] = df['Formatted_Date'].astype('object')
```

```python
# convert the date column to datetime format
df['Formatted_Date'] = pd.to_datetime(df['Formatted_Date'])

# convert the datetime format to the desired format
df['Formatted_Date'] = df['Formatted_Date'].dt.strftime('%Y-%m-%d')
```
